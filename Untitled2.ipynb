```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE

# Step 1: Load datasets (you can upload them directly to Colab or link to files)
# Upload your datasets first using Colab's file uploader if needed
cardio_data = pd.read_csv('/content/Cardiovascular_Disease_Dataset (1).csv')
diabetes_data = pd.read_csv('/content/Dataset of Diabetes .csv')
```

```python
# Step 2: Preprocess and merge the datasets
# Rename columns for consistency in merging
cardio_data.rename(columns={'age': 'AGE', 'gender': 'Gender'}, inplace=True)

# Handle NaN and non-numeric values in 'Gender' column
# Option 1: Assuming NaN means Female (0)
diabetes_data = diabetes_data.copy()  # Avoid chained assignment warning

# Fill missing Gender values with 'Female' (0)
diabetes_data['Gender'] = diabetes_data['Gender'].fillna(0)

# Convert Gender to numeric (M = 1, F = 0), ensuring that all remaining values are valid
diabetes_data['Gender'] = diabetes_data['Gender'].replace({'M': 1, 'F': 0})

# After replacing, convert to integers, but also coerce any remaining invalid entries
diabetes_data['Gender'] = pd.to_numeric(diabetes_data['Gender'], errors='coerce')

# Ensure that any remaining NaN values are filled with a default value, such as '0' for Female
diabetes_data['Gender'] = diabetes_data['Gender'].fillna(0)

# Convert the column to integer now that all invalid values have been handled
diabetes_data['Gender'] = diabetes_data['Gender'].astype(int)

# Ensure AGE and Gender columns are integers
cardio_data['AGE'] = cardio_data['AGE'].astype(int)
diabetes_data['AGE'] = diabetes_data['AGE'].astype(int)

# Continue with merging and the rest of the code...
```

```python
# Step 3: Select necessary columns for features and target labels
cardio_features = cardio_data[['AGE', 'Gender', 'serumcholestrol', 'restingBP', 'maxheartrate', 'target']]
diabetes_features = diabetes_data[['AGE', 'Gender', 'Chol', 'HbA1c', 'BMI', 'CLASS']]  # Added 'BMI'

# Merge both datasets on 'AGE' and 'Gender'
merged_data = pd.merge(cardio_features, diabetes_features, on=['AGE', 'Gender'], how='inner')
```

```python
# Step 4: Handle NaN values in the merged dataset
merged_data.loc[:, 'target'] = merged_data['target'].fillna(0)  # Assuming missing target means no CVD
merged_data.loc[:, 'CLASS'] = merged_data['CLASS'].fillna(0)    # Assuming missing CLASS means no Diabetes
```

```python
# Step 5: Split the data into features and targets
X = merged_data[['AGE', 'Gender', 'serumcholestrol', 'restingBP', 'Chol', 'HbA1c', 'maxheartrate', 'BMI']]  # Added 'maxheartrate' and 'BMI'
y_target = merged_data['target']  # Target for Cardiovascular Disease
y_class = merged_data['CLASS']    # Target for Diabetes

# Stratify the data by 'CLASS' to ensure balanced representation in the training set
X_train, X_test, y_train_target, y_test_target = train_test_split(X, y_target, test_size=0.3, random_state=42)
X_train_class, X_test_class, y_train_class, y_test_class = train_test_split(X, y_class, test_size=0.3, random_state=42, stratify=y_class)
```

```python
# Step 6: Apply SMOTE to balance the classes in the training data for Diabetes (CLASS)
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
X_train_resampled_class, y_train_resampled_class = smote.fit_resample(X_train_class, y_train_class)
```

```python
# Step 7: Train separate models for Cardiovascular Disease and Diabetes

# Train model for Cardiovascular Disease (target)
from sklearn.ensemble import RandomForestClassifier
rf_model_target = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model_target.fit(X_train, y_train_target)

# Train model for Diabetes (CLASS)
rf_model_class = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model_class.fit(X_train_resampled_class, y_train_resampled_class)
```

```python
# Step 8: Make predictions on the test set
y_pred_target = rf_model_target.predict(X_test)
y_pred_class = rf_model_class.predict(X_test_class)
```

```python
# Step 9: Evaluate the models

# Cardiovascular Disease (CVD) Classification Report
from sklearn.metrics import classification_report, confusion_matrix
print("Classification Report for Cardiovascular Disease (CVD):")
print(classification_report(y_test_target, y_pred_target, target_names=['No CVD', 'CVD']))

# Diabetes Classification Report
print("\nClassification Report for Diabetes:")
print(classification_report(y_test_class, y_pred_class, target_names=['No Diabetes', 'Diabetes']))
```

```python
# Confusion matrices for CVD and Diabetes
cm_cvd = confusion_matrix(y_test_target, y_pred_target)
cm_diabetes = confusion_matrix(y_test_class, y_pred_class)

# Plot the confusion matrix for CVD
import seaborn as sns
sns.heatmap(cm_cvd, annot=True, fmt="d", cmap="Blues", xticklabels=['No CVD', 'CVD'], yticklabels=['No CVD', 'CVD'])
plt.title('Confusion Matrix for Cardiovascular Disease')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# Plot the confusion matrix for Diabetes
sns.heatmap(cm_diabetes, annot=True, fmt="d", cmap="Oranges", xticklabels=['No Diabetes', 'Diabetes'], yticklabels=['No Diabetes', 'Diabetes'])
plt.title('Confusion Matrix for Diabetes')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
# Step 10: Feature Importance for Cardiovascular Disease
importances = rf_model_target.feature_importances_
features = X.columns

# Sort feature importance in descending order
indices = np.argsort(importances)[::-1]

plt.figure(figsize=(10, 6))
plt.title("Feature Importance in Predicting Cardiovascular Disease")
plt.bar(range(len(importances)), importances[indices], color="r", align="center")
plt.xticks(range(len(importances)), [features[i] for i in indices], rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Step 12: Take user input and predict (ensure correct order and feature names)
def get_user_input():
    age = int(input("Enter your age: "))
    gender = int(input("Enter your gender (1 for male, 0 for female): "))
    serum_chol = float(input("Enter your serum cholesterol level: "))  # This matches 'serumcholestrol'
    chol = float(input("Enter your Cholesterol level: "))  # This matches 'Chol'
    resting_bp = float(input("Enter your resting blood pressure: "))
    hba1c = float(input("Enter your HbA1c level: "))
    max_heartrate = float(input("Enter your maximum heart rate: "))  # New input for max heart rate
    bmi = float(input("Enter your Body Mass Index (BMI): "))  # New input for BMI

    # Ensure feature names are passed in the same order and match the training phase
    return pd.DataFrame([[age, gender, serum_chol, resting_bp, chol, hba1c, max_heartrate, bmi]],
                        columns=['AGE', 'Gender', 'serumcholestrol', 'restingBP', 'Chol', 'HbA1c', 'maxheartrate', 'BMI'])  # Added 'maxheartrate' and 'BMI'
```

```python
# Step 13: Make predictions for user input
user_input = get_user_input()

# Ensure that the user input has the same columns in the same order as during training
X_train_columns = ['AGE', 'Gender', 'serumcholestrol', 'restingBP', 'Chol', 'HbA1c', 'maxheartrate', 'BMI']  # Added 'maxheartrate' and 'BMI'
user_input = user_input[X_train_columns]  # Reorder columns if necessary

# Make predictions
cvd_prediction = rf_model_target.predict(user_input)[0]
diabetes_prediction = rf_model_class.predict(user_input)[0]

# Step 14: Display prediction results to the user
if cvd_prediction == 0:
    print("Prediction: No Cardiovascular Disease (CVD).")
else:
    print("Prediction: Cardiovascular Disease (CVD) detected.")

if diabetes_prediction == 0:
    print("Prediction: No Diabetes.")
else:
    print("Prediction: Diabetes detected.")
```
